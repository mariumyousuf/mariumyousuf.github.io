```python
import matplotlib.pyplot as plt
import numpy as np

# Data labels (questions)
categories = [
    'Fairness',
    'Stuck between two choices',
    'Clear & Straightforward',
    'Clear & Tricky',
    'Unclear',
    'Mix of Easy & Difficult',
    'Would like instructor to go over strategies'
]

# Data: [Yes, No] percentages
data = [
    [84.2, 15.8],
    [21.1, 57.9],
    [36.8, 63.2],       # 1 - 36.8
    [73.7, 26.3],       # 1 - 73.7
    [15.8, 84.2],       # 1 - 15.8
    [63.2, 36.8],       # 1 - 63.2
    [68.4, 31.6]
]

# Extract Yes and No values
yes_values = [item[0] for item in data]
no_values = [item[1] for item in data]

# X-axis positions
x = np.arange(len(categories))
width = 0.35

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(x - width/2, yes_values, width, label='Yes', color='green')
ax.bar(x + width/2, no_values, width, label='No', color='orange')

# Labels and titles
ax.set_ylabel('Percentage')
ax.set_title('Pre-Intervention Survey Results')
ax.set_ylim(0, 100)
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=30, ha='right')
ax.legend()

# Annotate bars with values
for i in range(len(categories)):
    ax.text(x[i] - width/2, yes_values[i] + 1, f'{yes_values[i]}%', ha='center', fontsize=8)
    ax.text(x[i] + width/2, no_values[i] + 1, f'{no_values[i]}%', ha='center', fontsize=8)

plt.tight_layout()
plt.show()
```
